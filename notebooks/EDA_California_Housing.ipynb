```python
# Importations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
import plotly.express as px
```

```python
# Creating the DataFrame
housing = fetch_california_housing()
features = pd.DataFrame(housing.data, columns=housing.feature_names)
target = pd.DataFrame(housing.target, columns=housing.target_names)
data = pd.concat([features, target], axis=1)
```

```python
# First Idea of the DataFrame
data.head()
```

```python
# Statistics data
data.describe()
```

```python
# General information about data (There is no Nulls and all the Features are float64)
data.info()
```

```python
# MedInc variable plot
sns.displot(data, x = "MedInc")
```

```python
# HouseAge variable plot
sns.displot(data, x = "HouseAge", kde=True)
```

```python
# Crear el grfico
sns.displot(data, x="AveRooms")

# Establecer los lmites del eje x entre 0 y 20
plt.xlim(0, 10)

# Mostrar el grfico
plt.show()
```

```python
fig = px.histogram(data, x="AveRooms")
fig.update_xaxes(range=[0, 10])
fig.show()
```
